# Базовая обработка изображений

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import imageio
```

```python
# Загрузка цветного изображения
img = imageio.imread('/content/9056_preview.v1.jpg')
img = img.astype(np.float32) / 255.
print(img.shape)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
img_red = np.copy(img[:, :, 0])
print(img_red.shape)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img_red) # cmap='gray'
```

# Оконная функция (функция фильтра)

```python
def my_filter(x):
    # Box blur
    return np.min(x)
```

```python
win_size = (11, 11) # размер окна

out_img = np.zeros(
    (img_red.shape[0] - win_size[0] + 1,
     img_red.shape[1] - win_size[1] + 1), 
     dtype=np.float32)

for i in range(out_img.shape[0]):
    for j in range(out_img.shape[1]):
          x = img_red[i:i+win_size[0], j:j+win_size[1]]
          out_img[i,j] = my_filter(x)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(out_img) # cmap='gray'
```
